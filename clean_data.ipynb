```python
import json
import os
import shutil

from pathlib import Path
import pandas as pd

from inaturalist import MARINE_CLASSES
```

```python
data_dir = Path("data")

train_dir = data_dir / "train"
val_dir = data_dir / "val"

train_json = data_dir / "train.json"
val_json = data_dir / "val.json"
```

```python
with open(train_json) as f:
    data = json.load(f)

    df = pd.DataFrame.from_dict(data["categories"])
    df = df.set_index(keys="id")
```

```python
df = df[df['class'].isin(MARINE_CLASSES)]
df
```

```python
marine_folders = df["image_dir_name"].to_list()
```

```python
marine_folders
```

```python
def filter_folders(base_directory):
    for folder_name in os.listdir(base_directory):
        # Get the full path of the folder
        folder_path = os.path.join(base_directory, folder_name)

        # Check if it is a directory and not in the list of folders to keep
        if os.path.isdir(folder_path) and folder_name not in marine_folders:
            # Delete the folder
            shutil.rmtree(folder_path)
```

```python
filter_folders(train_dir)
filter_folders(val_dir)
```
